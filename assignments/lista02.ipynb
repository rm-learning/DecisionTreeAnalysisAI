### **Questo** 2
Leia o artigo [A_comparative_study_of_decision_tree_ID3_and_C4.5.pdf](https://saiconference.com/Downloads/SpecialIssueNo10/Paper_3-A_comparative_study_of_decision_tree_ID3_and_C4.5.pdf) e responda:
- **Quais as diferenas entre os algoritmos de rvore ID3 e C4.5?**
  
  Segundo o artigo de HSSINA, MERBOUHA, EZZIKOURI e ERRITALI o algoritmo ID3 seleciona o melhor atributo com base nos conceitos de entropia e ganho de informao para desenvolver a rvore. Enquanto o algoritmo C4.5 age de maneira semelhante ao ID3, mas melhora alguns comportamentos do ID3, como a possibilidade de usar dados contnuos, usar valores desconhecidos (faltantes), capacidade de usar atributos com pesos diferentes e podar a rvore aps ser criada.

- **Como o algoritmo C4.5 lida com os atributos de entrada que so numricos?**
  
  Segundo o artigo, o algoritmo C4.5 lida com os atributos de entrada  partir de uma srie de passos. Primeiramente, ele ordena os valores do atributo em ordem crescente e elimina os valores repetidos, depois considera cada valor como um possvel ponto de corte para dividir os dados em dois subconjuntos: um com valores menores ou iguais ao ponto de corte e outro com valores maiores. Ento, ele calcula o ganho ou a razo de ganho para cada diviso e escolhe a que maximiza o critrio e, na sequncia, usa o valor escolhido como um teste binrio na forma de atributo <= valor ou atributo > valor para gerar um ramo na rvore de deciso.
