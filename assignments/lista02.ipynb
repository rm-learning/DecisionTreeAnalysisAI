### **Questão** 2️⃣
Leia o artigo “[A_comparative_study_of_decision_tree_ID3_and_C4.5.pdf](https://saiconference.com/Downloads/SpecialIssueNo10/Paper_3-A_comparative_study_of_decision_tree_ID3_and_C4.5.pdf)” e responda:
- **Quais as diferenças entre os algoritmos de árvore ID3 e C4.5?**
  
  Segundo o artigo de HSSINA, MERBOUHA, EZZIKOURI e ERRITALI o algoritmo ID3 seleciona o melhor atributo com base nos conceitos de entropia e ganho de informação para desenvolver a árvore. Enquanto o algoritmo C4.5 age de maneira semelhante ao ID3, mas melhora alguns comportamentos do ID3, como a possibilidade de usar dados contínuos, usar valores desconhecidos (faltantes), capacidade de usar atributos com pesos diferentes e podar a árvore após ser criada.

- **Como o algoritmo C4.5 lida com os atributos de entrada que são numéricos?**
  
  Segundo o artigo, o algoritmo C4.5 lida com os atributos de entrada à partir de uma série de passos. Primeiramente, ele ordena os valores do atributo em ordem crescente e elimina os valores repetidos, depois considera cada valor como um possível ponto de corte para dividir os dados em dois subconjuntos: um com valores menores ou iguais ao ponto de corte e outro com valores maiores. Então, ele calcula o ganho ou a razão de ganho para cada divisão e escolhe a que maximiza o critério e, na sequência, usa o valor escolhido como um teste binário na forma de “atributo <= valor” ou “atributo > valor” para gerar um ramo na árvore de decisão.
